# **Movie Recommendation System**

**Recommendation System** is an artificial intelligence or AI algorithm, usually associated with machine learning, that uses Big Data to suggest or recommend additional products to consumers.

**Collaborative Filtering** is a family of algorithms where there are multiple ways to find similar users or items and multiple ways to calculate rating based on ratings of similar users.

**Content-based Filtering** uses item features to recommend other items similar to what the user likes, based on their previous actions or explicit feedback.

-------------

## **Objective**

Develop a basic recommendation system by recommending movies most similar to a particular movie.

The dataset encompasses diverse information about movies, including budget, popularity, release date, revenue, runtime, voting details, keywords, production details, and cast and crew information.

## **Import Library**

```python
import pandas as pd
import numpy as np
```

## **Import Data**

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

## **Data Preprocessing**

```python
df_features = df[['Movie_Genre', 
                  'Movie_Keywords',
                  'Movie_Tagline',
                  'Movie_Cast',
                  'Movie_Director']].fillna('')
df_features.shape
df_features
```

## **Feature Variables (X)**

```python
X = df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords'] + ' ' + df_features['Movie_Tagline'] + ' ' + df_features['Movie_Cast'] + ' ' + df_features['Movie_Director'] 
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

## **Model Evaluation**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(X)
```

```python
sim_score
```

```python
sim_score.shape
```

## **Prediction**

```python
favmovname = input('enter fav movie =')
```

```python
allmovtitlelist = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
movrec = difflib.get_close_matches(favmovname, allmovtitlelist)
print(movrec)
```

```python
closematch = movrec[0]
print(closematch)
```

```python
indexofclosematch = df[df.Movie_Title == closematch]['Movie_ID'].values[0]
print(indexofclosematch)
```

```python
recscore = list(enumerate(sim_score[indexofclosematch]))
print(recscore)
```

```python
len(recscore)
```

```python
#sort movies based on simscore

sortedsimmov = sorted(recscore, key = lambda x:x[1], reverse = True)
print(sortedsimmov)
```

```python
print('top 30 movies suggested for you : \n')

i = 1
for movie in sortedsimmov:
    index = movie[0]
    titlefromindex = df[df.index==index]['Movie_Title'].values[0]
    if (i<31):
        print(i,'.',titlefromindex)
        i+=1
```
